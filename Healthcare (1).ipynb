```python
# 1. Medical Condition Frequency
condition_counts = df['Medical Condition'].value_counts()
print("Medical Condition Frequency:\n", condition_counts)
```

```python
Gender Distribution in Medical Conditions
gender_condition = df.groupby(['Gender', 'Medical Condition']).size()
print("\nGender Distribution in Medical Conditions:\n", gender_condition)
```

```python
Average Billing Amount by Insurance Provider
avg_billing_by_insurance = df.groupby('Insurance Provider')['Billing Amount'].mean()
print("\nAverage Billing Amount by Insurance Provider:\n", avg_billing_by_insurance)
```

```python
Average Hospital Stay (Discharge Date - Admission Date) by Admission Type
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Hospital Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
avg_stay_by_admission = df.groupby('Admission Type')['Hospital Stay'].mean()
print("\nAverage Hospital Stay by Admission Type:\n", avg_stay_by_admission)
```

```python
Medication Effectiveness (Test Results by Medication)
medication_effectiveness = df.groupby(['Medication', 'Test Results']).size()
print("\nMedication Effectiveness:\n", medication_effectiveness)
```

```python
Age Distribution by Medical Condition
age_condition = df.groupby('Medical Condition')['Age'].mean()
print("\nAverage Age by Medical Condition:\n", age_condition)
```

```python
Hospital Analysis: Admissions by Type per Hospital
hospital_admissions = df.groupby(['Hospital', 'Admission Type']).size()
print("\nHospital Admissions by Type:\n", hospital_admissions)
```

```python
billing_by_condition = df.groupby('Medical Condition')['Billing Amount'].mean().sort_values(ascending=False)
print(billing_by_condition)
```

```python
blood_type_distribution = df['Blood Type'].value_counts()
print(blood_type_distribution)
```

```python
Age Distribution
plt.figure(figsize=(8,6))
sns.histplot(df['Age'], kde=True, bins=10, color='blue')
plt.title('Age Distribution of Patients', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y='Medical Condition', data=df, order=df['Medical Condition'].value_counts().index, hue='Medical Condition', palette='Set2', dodge=False)
plt.title('Frequency of Medical Conditions', fontsize=16)
plt.xlabel('Count', fontsize=12)
plt.ylabel('Medical Condition', fontsize=12)
plt.legend([], [], frameon=False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
billing_by_condition.plot(kind='bar', color='green')
plt.title('Average Billing Amount by Medical Condition', fontsize=16)
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Average Billing Amount', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
df['Length of Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
length_of_stay = df.groupby('Medical Condition')['Length of Stay'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
length_of_stay.plot(kind='bar', color='orange')
plt.title('Average Length of Stay by Medical Condition', fontsize=16)
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Average Length of Stay (days)', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,8))
blood_type_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Blood Type Distribution', fontsize=16)
plt.ylabel('')  # Hide the y-label for a cleaner look
plt.show()
```

```python
# Scatter Plot of Length of Stay by Medical Condition
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Medical Condition', y='Length of Stay', data=df, hue='Medical Condition', palette='Set2', s=100)
plt.title('Scatter Plot: Length of Stay by Medical Condition', fontsize=16)
plt.xlabel('Medical Condition', fontsize=12)
plt.ylabel('Length of Stay (days)', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Medical Condition', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
plt.figure(figsize=(8, 8))
admission_counts = df['Admission_Type'].value_counts()
plt.pie(admission_counts, labels=admission_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Admission Types')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.legend(title='Admission_Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
# Create age groups
bins = [0, 20, 40, 60, 80, 100]
labels = ['0-20', '21-40', '41-60', '61-80', '81-100']
df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

sns.boxplot(data=df, x='Age_Group', y='Billing_Amount')  # Removed palette argument
plt.title('Billing Amount Distribution by Age Group')
plt.ylabel('Billing Amount ($)')
plt.xlabel('Age Group')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', color="pink")
plt.title('Distribution of Patients by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Billing_Amount'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Billing Amounts')
plt.xlabel('Billing Amount ($)')
plt.ylabel('Frequency')
plt.axvline(df['Billing_Amount'].mean(), color='red', linestyle='dashed', linewidth=1, label='Mean')
plt.legend()
plt.tight_layout()
plt.show()
```
